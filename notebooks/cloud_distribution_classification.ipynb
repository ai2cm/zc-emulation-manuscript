```python
import xarray as xr
import matplotlib.pyplot as plt
import numpy as np

import common
```

```python
plt.style.use("tableau-colorblind10")
```

```python
data = xr.open_dataset("fig_data/cloud_day15_classifier_comparison.nc")
```

```python
plt.figure(figsize=(6, 4), dpi=120)
bins = np.logspace(-1, 4, 50)
for k, v in data.items():
    v.plot.hist(bins=bins, histtype="step", xscale="log", yscale="log", label=k)
plt.title("Cloud water distribution - July 15")
plt.xlabel("mg/kg")
plt.ylabel("count")
plt.xlim(1e-1, 1e4)
plt.legend(loc="lower left")
common.savefig("cloud_distribution_classification")
plt.show()
```
